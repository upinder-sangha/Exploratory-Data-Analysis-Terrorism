##Reading the dataset


```
dataset = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='latin-1')
df = pd.DataFrame(dataset)
```

```
pd.set_option('display.max_columns', None)
dataset
```

##Simplifying the dataset
Since there are too many columns and we don't need all of them in our analysis, so we will extract only the ones we need

```
df = df[["iyear","imonth","extended","country","country_txt","region","region_txt","specificity","vicinity","crit1","crit2","crit3","doubtterr","multiple","success","suicide","attacktype1","attacktype1_txt","targtype1","targtype1_txt","weaptype1","weaptype1_txt","nkill","nwound","property","ishostkid","ransom"]]
```

#Analysis of dataset using graphs and figures

##Line graph showing the trend of number of cases of terrorism over the years

```
cases_table = pd.pivot_table(data=df, index='iyear',values='region', aggfunc=np.count_nonzero)
cases_table.plot.line(figsize=(15,5), legend=None, xlabel= "Year", ylabel= "Number of cases")
```

##Line graph showing the trend of number of cases of terrorism over years for different regions

```
cases_table2 = pd.pivot_table(data=df, index='iyear', columns='region_txt', values='region', aggfunc=np.count_nonzero)
cases_table2.plot.line(figsize=(15,5)).legend(title=None)
```

##Analyzing the number of cases of terrorism in different countries and comparing them using bar graph

```
cases_table3 = pd.pivot_table(data=df, index='country_txt', columns=None, values='region', aggfunc=np.count_nonzero)
cases_table3 = cases_table3.sort_values('region', ascending=False)
cases_table3 = cases_table3.head(50)
cases_table3.plot.bar(figsize=(15,5),legend=None)
```

##Comparing the number of cases of terrorism in different regions of the world

```
nkilltable = pd.pivot_table(data=df, index='region_txt', values='region', aggfunc=np.count_nonzero)
nkilltable.plot.pie(subplots = True, figsize=(8,8), legend = None)
```

##Analyzing the types of terrorism assaults and their success rates 
This data is useful especially in preventing furthur assaults of these types

```
attack_type_table = pd.pivot_table(data=df, index='attacktype1_txt', columns='success', values='region', aggfunc=np.count_nonzero)
attack_type_table = attack_type_table.sort_values(1, ascending=False)
attack_type_table.plot.bar(figsize=(15,5))
```

##A bar graph showing the types of people targeted by the terrorist assault 

```
nkilltable = pd.pivot_table(data=df, index='targtype1_txt', columns='success', values='region', aggfunc=np.count_nonzero)
nkilltable = nkilltable.sort_values(1, ascending=False)
nkilltable.plot.bar(figsize=(15,5))
```

##A bar graph comparing the types of weapon used in terrorist attacks

```
weapon_table = pd.pivot_table(data=df, index='weaptype1_txt', columns=None, values='nkill', aggfunc=np.sum)
weapon_table = weapon_table.sort_values('nkill', ascending=False)
weapon_table.plot.bar(figsize=(15,5))
```

##A bar graph showing the average number of deaths vs the type of weapon used in the attack


```
weapon_table2 = pd.pivot_table(data=df, index='weaptype1_txt', columns=None, values='nkill', aggfunc=np.mean)
weapon_table2 = weapon_table2.sort_values('nkill', ascending=False)
weapon_table2.plot.bar(figsize=(15,5))
```
